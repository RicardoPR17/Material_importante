# Task: crear un listado con unos nombres, despues crear un metodo que recibe el listado y el numero n.El metodo debe imprimir y devolver los nombres del listado que tienen este tamano (numero n)

```python
#
#
#
#
#
#
#
# el algoritmo
# 1. definir el listado de los nombres
names = ["Diego","Felipe","Maria","Ana","Lida","Luis","Alfredo","Alejandro"
,"Daniel","Dario","Rodrigo","Jose","Maria","Raul","Laura","Daniela","Alejandra",
"Sofia","Elena"]
# 2. definir un metodo que reciba dos variables: el listado de los nombres y un nomero
# 3. iterar por los elementos del listado y medir los tamanos de los elemento. Si el tamano
# es igual que n en este caso imprimir el nombre
def get_name_by_size(names,n):
  print("="*64)
  print("n es:",n)
  for name in names:
    if(len(name)==n):
      print(name, end=" ")
  print()
# 4. llamar el metodo
for i in range(2,10):
  get_name_by_size(names,i)

```

# Task: Para la tarea anterio, que hacer para mostrar el ultimo que cumple?

```python
def get_name_by_size_last(names,n):
  print("="*64)
  print("n es:",n)
  last_name = ""
  for name in names:
    if(len(name)==n):
      last_name = name
  print(last_name)


# 4. llamar el metodo
for i in range(2,10):
  get_name_by_size_last(names,i)
```

Task: Para la tarea anterio, queremos recibir otro numero y este numero va a ser la posicion entre los valores que cumplen

```python
def get_name_by_size_by_position(names,n,pos):
  print("="*64)
  print("n es:",n)
  the_name = ""
  contador = 0
  for name in names:
    if(len(name)==n):
      contador +=1
      if(contador == pos):
        the_name = name
        break
  print(the_name)


get_name_by_size(names,5)# Diego Maria Dario Maria Laura Sofia Elena
get_name_by_size_by_position(names,5,3)# Dario
get_name_by_size_by_position(names,5,5)# Laura
```

```python
def print_digits(n):
  print(n)
  text = ""
  while(n>=10):
    text  += "," + str(n%10)
    n = int(n/10)
  text += "," + str(n)
  print(text)
n = 987654
print_digits(n)
```

```python
def print_digits_list(n):
  print(n)
  list= []
  while(n>=10):
    list.append(str(n%10))
    n = n//10
  list.append(str(n))
  a = ", ".join(list)
  print(a)
n = 987654
print_digits_list(n)
```

```python
def print_digits_inverse(n):
  print(n)
  while(n>=10):
    print(n%10, end=',')
    n = int(n/10)
  print(n)
n = 987654
print_digits_inverse(n)
```

```python
def print_digits_text(n):
  a = str(n) # covierte a un string o texto
  # itera desde el final del texto hasta el inicio
  for i in range(len(a)-1,-1,-1):
    print(a[i],end=',')
  print()
  for i in range(0,len(a),+1):
    print(a[i],end=',')

n = 12345678
print_digits_text(n)
```

# Task
Create a method that receives the
number n and returns the sum of all odd numbers before it
example n=5 -> returns 1+3+5= 9

```python
def sum_odds_before(n):
  sum_temp = 0
  for i in range(1,n+1):
    if(i%2==1):
      sum_temp +=i
      if(i!=n and i!=n-1):
        print(i,end='+')
      else:
        print(i,end='=')
  print(sum_temp)

sum_odds_before(9)
sum_odds_before(10)
```

```python
def sum_of_odds(n):
  sum=0
  for i in range(1,n+1,+2): # i = 1 3 5 7 ...
    sum+=i    # sum = sum + i
  return sum

def  sum_of_odd_if(n):
  sum=0
  for i in range(n+1):
    if i%2!=0: # si dividimos i por 2 el residuo no debe estar igual cero
      sum+=i
  return sum

print(sum_of_odds(5))
# print(sum_of_odds(6))

print(sum_of_odd_if(5))
# print(sum_of_odd_if(6))
```

# Task 2 list = [1,2,3,4,5,6,7,8,9], Make a method that returns min and max Do not use min, max, nor sort!

```python
list = [30,2,12,4,55,6,7,98,91]
def min_max(list):
  tmin = tmax = list[0]
  for x in list:
    if x<tmin:
      tmin = x
    if x>tmax:
      tmax = x
  return [tmin,tmax]
  # return min,max # (min,max)-> tuple

print(type(min_max(list)))
print(min_max(list))

def min_max_easy(list):
  tmin = min(list)
  tmax = max(list)
  return [tmin,tmax]

def min_max_by_sort(list):
  list.sort()
  return [list[0],list[-1]]

print(min_max_easy(list))
print(min_max_by_sort(list))
```
