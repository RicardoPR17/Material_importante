Sets

Set Items

Set items are unordered, unchangeable, and do not allow duplicate values.
Unordered

Unordered means that the items in a set do not have a defined order.

Set items can appear in a different order every time you use them, and cannot be referred to by index or key.
Unchangeable

Sets are unchangeable, meaning that we cannot change the items after the set has been created.

```python
myset = {"Colombia", "Peru", "Mexico"}
print(myset)
myset = {"Colombia", "Peru", "Mexico","Colombia"} # los duplicados se ignoran
print(myset)
print(len(myset))
set1 = {"abc", 34, True, 40, "male"}
print(type(myset)) #<class 'set'>
thisset = set(("apple", "banana", "cherry","banana")) # note the double round-brackets
print(thisset)
for x in thisset:
  print(x, end=",")
print("banana" in thisset)
thisset.add("orange")
print(thisset)
tropical = {"pineapple", "mango", "papaya","cherry"}
thisset.update(tropical)
print(thisset)
mylist = ["kiwi", "maracuya"]
thisset.update(mylist)
print(thisset)
```

```python
# If the item to remove does not exist, remove() will raise an error.
print(thisset.remove("orange"))
print(thisset)
thisset.discard("pineapple") # descarta
thisset.discard("test") # descarta
print(thisset)
x = thisset.pop()
print(x)
print(thisset)
```

```python
# Class in Python

class Employee:
  # aca se definen las variablesque pertenecen a la clase (static)
  value=10
  def __init__(self,name,salary, private_variable):
    # aca se definen las propeidades de la clase
    self.name = name
    self.salary = salary
    self.__private_variable = private_variable

  def assign_salary(self,salary):
    self.salary = salary

  def __str__(self): # java -> override toString()
    return self.name + " " + str(self.salary)

employee1 = Employee("Daniel", 4000000,10)
print(employee1.name)
print(employee1.value)
employee1.value += 3
print(employee1.value)
employee1.assign_salary(4000100)
# print(employee1.__private_variable) # private
employee2 = Employee("Sofia", 6000000,15)
print(employee2.name)
print(employee2.value)
print(Employee.value)
```

```python
class Person:
  def __init__(self): # constructor
    self.name = "Name"
    self.age = 18
  def __str__(self): # override toString
    return self.name +" "+ str(self.age)
  def change_name(self, name):
    self.name = name

p1 = Person()
print(p1)
p1.change_name("Lida")
print(p1.name)
print(p1.age)
```
